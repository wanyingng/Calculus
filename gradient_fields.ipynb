```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plot a trigonometric function of two variables, $X$ and $Y$

```python
# Create a set of 51 points from -3 to 3
xx = np.linspace(-3,3,51)

X, Y = np.meshgrid(xx, xx)

# Create function matrix
Z = np.sin(X * np.cos(Y))

# Compute the two partial derivatives of the multivariable function
gx, gy = np.gradient(Z)

# Visualize the mathematical function
plt.contourf(xx, xx, Z, 40)
```

We transformed a mathematical expression into an engaging visualization.

Next, we'll plot the gradient, which represents all the partial derivatives of a multivariable function.

```python
plt.contourf(xx, xx, Z, 40)
plt.quiver(xx, xx, gx, gy)
```

Here, we have added a visual represention of all the gradients, illustrated by small arrows.

In this visualization, bluish-purple hues indicate negative function values, while yellow represents positive values.

Notably, the arrows—depicting the gradients—point toward the regions where the function attains positive values (the yellow areas).

Let's add some parameters to this mathematical function.

```python
f1 = 4
f2 = 0.8

# Create function matrix
Z = np.sin(f1 * X * np.cos(f2 * Y))

# Compute the two partial derivatives of the multivariable function
gx, gy = np.gradient(Z)

# Visualize the mathematical function
plt.contourf(xx, xx, Z, 40)
plt.quiver(xx, xx, gx, gy)
```

The parameters `f1` and `f2` control the frequency of the sine and cosine waves, making the plot appear busier as their values increase. Higher parameter values mean higher frequency, resulting in more repetitions of these sinusoidal patterns.

Another parameter we'll be adding to this mathematical function is the *phase offset*. 

```python
# Define the phase offsets
p1 = 0.6
p2 = -0.77

# Create function matrix
Z = np.sin(f1 * X * np.cos(f2 * Y + p2) + p1)

# Compute the two partial derivatives of the multivariable function
gx, gy = np.gradient(Z)

# Visualize the mathematical function
plt.contourf(xx, xx, Z, 40)
plt.quiver(xx, xx, gx, gy)
```

The phase offsets introduce a shift.

Finally, let's explore how to adjust the color scheme of this visualization.

```python
# Adjust the color scheme
plt.contourf(xx, xx, Z, 40, cmap="pink")
plt.quiver(xx, xx, gx, gy)
```
